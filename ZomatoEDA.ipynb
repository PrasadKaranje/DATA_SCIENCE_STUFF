```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
## to check data types
final_df.dtypes
```

```python
fn_df=final_df
```

```python
country_names=fn_df.Country.value_counts().index
```

```python
country_val=fn_df.Country.value_counts().values
```

```python
##Pie Chart
plt.pie(country_val, labels=country_names)
```

```python
##Pie Chart of top threee
plt.pie(country_val[:3], labels=country_names[:3],autopct='%1.1f%%')
```

```python
ratings=final_df.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,10)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,10)
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['purple','red','orange','yellow','green','green'])
```
